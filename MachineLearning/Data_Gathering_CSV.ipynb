### DATA GATHERING WITH CSV
### tackling the challenges that we might face while data gathering using CSV


```python
import pandas as pd
```

```python
pd.read_csv('Titanic.csv')
```

### TO READ A TSV ( TAB SEPERATED VALUE) WE NEED TO USE CSV---> SEP FEATURE
### ---> CSV SEP FEATURE ---> SEPERATION FEATURE
### CSV ---> amit,28,M ----> SEPERATION = ','
### TSV ---> amit\t28\tM --> SEPERATION = '\t'

```python
pd.read_csv('test_Arabic_tweets_negative_20190413.tsv', sep = '\t')
```

### header = 1 removes the first row which is column names.
### NAMES PARAMETER ---> Names is going to take a list of names that we can provide to our colmnu names
### INDEC_COL ---> TO KEEP ONE OF COLUMN AS AN INDEX

```python
pd.read_csv('Titanic.csv', header = 1 ,names = [ 'P_Id', 'Pclass', 'P_Name', 'Gender', 'Age', 'SibSp', 'Parch', 'Ticket', 'Fare', 'Cabin', 'Embarked' ], index_col = 'P_Id')
```

### Use_Cols ---> WHAT COLUMNS WE SHOULD USE TO SHOW OUT ALL THE COLUMNS WE ARE HAVING IN THE DATAFRAME
### Use List to mention all the columns names that are required

```python
pd.read_csv('student_depression_dataset.csv', usecols = ['id','Gender','Age','City','Profession','Academic Pressure','CGPA','Study Satisfaction','Sleep Duration','Dietary Habits','Degree','Have you ever had suicidal thoughts ?','Work/Study Hours','Financial Stress','Family History of Mental Illness','Depression'])
```

### SQUEEZE ---> DataFrame ---> Series ---> Squeeze ---> True ---> Combination of USECOLS and squeeze ---> True

```python
pd.read_csv('Titanic.csv',  usecols = ['Name']).squeeze()
```

### nrows ---> Helps us with fixed number of rows out of the complete table

```python
pd.read_csv('student_depression_dataset.csv', nrows = 100)
```

### ENCODING --> MAKE SURE THAT THE DATA WE ARE READING IS IN UTF-8 FORMAT
# ENCODING VALUE = 'LATIN-1'
### ENCODING ---> WHILE WORKING WITH THE NUMERICAL KINDOF DATA --> UTF-8 IS USED
## WHEN THE DATA IS IN PRESENT IN THE TEXT FORMAT OR ANY OTHER NON NUMERICAL DATA 
### ---> THEN WE HAVE TO MANUALLY KEEP THE ENCODING TO --> UTF-8 -> TO READ THE CONTENT

```python
pd.read_csv('Titanic.csv', encoding = 'latin-1')
```

### Dtypes ---> it can assign initially only with a specific data type to a column 
### ---> dtypes = {'col_1':int,'col_2':float}

```python
pd.read_csv('student_depression_dataset.csv', dtype = {'Age': int, 'Work Pressure': int})
```

## parse_date --> Convert the date into appropriate readale format

### CONVERTERS ---> 
#### BRING THE TRANSFORMATION ON A CONVERTER (SPECIFIC VALUE) IN A (COLUMN) & THIS CONVERTER IS A FUNCTION

```python
def station(value):
    if value == 'Q':
        return 'Queenstown'
    else:
        return value

pd.read_csv('Titanic.csv', converters = {'Embarked': station})
```

### na_values ---> helps with mentioning what values to be considered as missing values
### na_values = [ '/' , ':' , 'NO' ]

```python
pd.read_csv('Titanic.csv', na_values = ['NO', 'NaN' ])
```
