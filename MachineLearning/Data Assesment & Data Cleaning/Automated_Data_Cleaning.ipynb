# AUTOMATED DATA CLEANING

## Types of missing data :-
### 
### 1. Missing Completely At Random (MCAR)
### 2. Missing At Random (MAR)
### 3. Missing Not At Random (MNAR)

```python
import pandas as pd
```

```python
df = pd.read_csv('unclean_smartwatch_health_data.csv')
```

```python
df.head()
```

```python
df[df['Heart Rate (BPM)'].isnull() == True]
```

### There's No Pattern between the columns that signifies the Missing Values in 'Heart Rate (BPM). This Means that It is an example of Missing Completely At Random (MCAR)

```python
df['Heart Rate (BPM)'].fillna(df['Heart Rate (BPM)'].median(), inplace = True)
```

```python
df['Heart Rate (BPM)'].isnull().sum()
```

```python
df[df['Blood Oxygen Level (%)'].isnull() == True]
```

```python
df = df.drop(columns = ['User ID'])
```

```python
df.head()
```

```python
df['Blood Oxygen Level (%)'].fillna(df['Blood Oxygen Level (%)'].mean(),inplace = True)
```

```python
df.info()
```

```python
df[df['Step Count'].isnull() == True]
```

### Forward Filling --- Backward Filling Strategy
#### --> Filling the missing value from previous value is forward filling
#### --> Filling the missing value from upcoming value is backward filling

```python
df['Step Count']=df['Step Count'].fillna(method = 'ffill',)
```

```python
df.info()
```

### Object ---> Numeric Data ----> pandas ---> to_numeric()

```python
df['Sleep Duration (hours)'] = pd.to_numeric(df['Sleep Duration (hours)'], errors = 'coerce')
```

```python
df['Sleep Duration (hours)'].dtype
```

```python
df['Sleep Duration (hours)'].isnull().sum()
```

```python
df['Sleep Duration (hours)'] = df['Sleep Duration (hours)'].fillna(df['Sleep Duration (hours)'].mean())
```

```python
df['Sleep Duration (hours)'].isnull().sum()
```

### For Categorical Columns ---> We can use MODE function

```python
df.info()
```

```python
df['Activity Level'].mode()
```

```python
df['Activity Level'] = df['Activity Level'].fillna('Seddentary')
```

```python
df['Activity Level'].isnull().sum()
```

### constant imputation ---> when you want to film missing values with constant values like 'unknown', '-999','0'.

```python
df.info()
```

```python
df['Stress Level'] = df['Stress Level'].replace('Very High', 8)
```

```python
df.info()
```

```python
df['Stress Level'].value_counts()
```

```python
df['Stress Level'] = pd.to_numeric(df['Stress Level'], errors = 'coerce')
```

```python
df['Stress Level'].mode()
```

```python
df['Stress Level'] = df['Stress Level'].fillna(5)
```

```python
df['Stress Level'].value_counts()
```

```python
df['Stress Level'].isnull().sum()
```

```python
df.info()
```

```python
df['Heart Rate (BPM)'] = df['Heart Rate (BPM)'].astype('float32')
df['Blood Oxygen Level (%)'] = df['Blood Oxygen Level (%)'].astype('float32')
df['Step Count'] = df['Step Count'].astype('float32')
df['Sleep Duration (hours)'] = df['Sleep Duration (hours)'].astype('float32')
df['Stress Level'] = df['Stress Level'].astype('float32')
```

```python
df.info()
```

```python
df.to_csv('clean_smartwatch_health_data.csv', index = False)
```
