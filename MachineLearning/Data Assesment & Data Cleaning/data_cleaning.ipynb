# DATA ASSESMENT

## What all different results to come up with?
### Reduce the datatype size to save memory such as float64 --> float32
### 1. Remove Irrelevent Column Such as USER ID
### 2. Remove missing values if they are 5% or less then 5%
### 3. Fill Missing values with Mean/Mode/Median Usually Mean --- Fillna()
###

```python
df['Heart Rate (BPM)'].isnull().sum()
```

```python
df['Heart Rate (BPM)'].value_counts()
```

```python
df['Heart Rate (BPM)'].max()
```

```python
df[df['Heart Rate (BPM)'] == 296.5939695131042]
```

```python
df['Heart Rate (BPM)'].plot(kind = 'box')
```

```python
# df.dropna(subset = [col_1, COL_2, COL_3, etc], inplace = True)
df.dropna(subset = ['Heart Rate (BPM)'], inplace = True)
```

```python
df['Heart Rate (BPM)'].isnull().sum()
```

```python
df.columns
```

```python
df = df[['Heart Rate (BPM)', 'Blood Oxygen Level (%)', 'Step Count',
       'Sleep Duration (hours)', 'Activity Level', 'Stress Level']]
```

```python
df.head()
```

```python
df['Heart Rate (BPM)'] = df['Heart Rate (BPM)'].astype('float32')
```

```python
df['Blood Oxygen Level (%)'] = df['Blood Oxygen Level (%)'].astype('float32')
```

```python
df['Step Count'] = df['Step Count'].astype('float32')
```

```python
df.info()
```

```python
df['Blood Oxygen Level (%)'].isnull().sum()
```

```python
df['Blood Oxygen Level (%)'].mean()
```

```python
df['Blood Oxygen Level (%)'].fillna(df['Blood Oxygen Level (%)'].mean(), inplace = True)
```

```python
df['Blood Oxygen Level (%)'].isnull().sum()
```

```python
df.shape
```

```python
df.dropna(subset = ['Step Count'], inplace = True)
```

```python
df['Step Count'].isnull().sum()
```

```python
df['Sleep Duration (hours)'].value_counts()
```

```python
df = df[df['Sleep Duration (hours)'] != 'ERROR']
```

```python
df['Sleep Duration (hours)'].value_counts()
```

```python
df['Sleep Duration (hours)'].isnull().sum()
```

```python
df['Sleep Duration (hours)'] = df['Sleep Duration (hours)'].astype('float32')
```

```python
df['Sleep Duration (hours)'].mean()
```

```python
df['Sleep Duration (hours)'].fillna(6.5068035, inplace = True)
```

```python
df.shape
```

```python
df.head()
```

```python
df['Stress Level'].value_counts()
```

```python
df = df[df['Stress Level'] != 'Very High']
```

```python
df['Stress Level'].value_counts()
```

```python
df['Stress Level'].isnull().sum()
```

```python
df['Stress Level'] = df['Stress Level'].astype('float32')
```

```python
df['Stress Level'].mean()
```

```python
df['Stress Level'].fillna(5.0, inplace = True)
```

```python
df['Stress Level'].value_counts()
```

```python
df['Activity Level'].value_counts()
```

```python
df['Activity Level'].isnull().sum()
```

```python
df.dropna(subset = ['Activity Level'], inplace = True)
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df.to_csv('new_unclean_smartwatch_health_data.csv', index = False)
```
