## There's an order in Education field this is why we will use ordinal encoding
#### tertiary > Secondary > Primary > Unknown

# Input data must be in 2D Format
###
### Ordinal Encoder ---> Education
### Label Encoder ---> Target
### Nominal Encoder ---> Married, ... 


# Ordinal Encoder

```python
ordinal_data = bank[['Education']]
```

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(categories = [['unknown', 'primary', 'secondary', 'tertiary' ]])

ordinal_data = oe.fit_transform(ordinal_data)
```

```python
ordinal_data
```

```python
target_data = bank[['TARGET']]
# , 'Default', 'Housing', 'Loan' 
```
