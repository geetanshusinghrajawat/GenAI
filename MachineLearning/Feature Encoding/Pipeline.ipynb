```python
import pandas as pd
```

```python
df = pd.read_csv('Bank_Marketing_Original_Data.csv')
```

```python
df.head()
```

# STEP-1
### Data Gathering

```python
Numeric_Data = ['Age','Balance','Day','Duration', 'Campaign','Pdays','Previous']
Ordinal_Data = ['Education']
Nominal_Data = ['Job', 'Marital', 'Default','Housing','Loan','contact','Month','poutcome']
```

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder, StandardScaler
```

# STEP-2
### Seperate Data into Multiple Pipelines

```python
Numeric_Pipeline = Pipeline(steps = [('impute', SimpleImputer(strategy = 'mean')),('scaling', StandardScaler())])

Ordinal_Pipeline = Pipeline(steps = [('impute',SimpleImputer(strategy = 'most_frequent')),('encoding',OrdinalEncoder(categories = [['unknown','primary','secondary','tertiary']])),('scaling',StandardScaler())])

Nominal_Pipeline = Pipeline(steps = [('impute',SimpleImputer(strategy = 'most_frequent')),('encoding',OneHotEncoder(handle_unknown = 'ignore'))])
```

# Step 3
### Combine Pipeline transformers with their respective data

```python
from sklearn.compose import ColumnTransformer
```

```python
features = ColumnTransformer(transformers = [('Numeric Transformation', Numeric_Pipeline,Numeric_Data),('Ordinal Transformation', Ordinal_Pipeline,Ordinal_Data),('Nominal Transformation',Nominal_Pipeline,Nominal_Data)])
```

# Step 4
### Create an model with respective transformers and along with their Algorithm

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = Pipeline(steps = [('features',features),('classifier',LogisticRegression())])
```

```python
model
```

# Step 5
### Train-test Split of DATA & Encoding of the OUTPUT column Data.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
X = df.iloc[:,:-1]
Y = df['TARGET']
```

```python
le = LabelEncoder()
Y = le.fit_transform(Y)
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2)
```

# Step 6 (FINAL STEP)
### Model Training & Getting Predictions From Model

```python
model.fit(X_train, Y_train)
```

```python
Y_predict = model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(Y_predict,Y_test)
```

### Using Different ML Model

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model = Pipeline(steps = [('features',features),('classifier',RandomForestClassifier())])
```

```python
model.fit(X_train,Y_train)
```

```python
Y_predict = model.predict(X_test)
accuracy_score(Y_predict,Y_test)
```
